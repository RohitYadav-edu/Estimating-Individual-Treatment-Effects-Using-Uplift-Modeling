```python
# Inspecting the large file

path = '/Users/rohityadav/Desktop/Git Projects/ml-uplift-modeling-criteo/data/criteo-uplift-v2.1.csv'
df_large = pd.read_csv(path)
df_large.head()
```

```python
df_large['treatment'].value_counts()
```

```python
df_large.groupby('treatment')['visit'].mean()
```

```python
df_large.info()
```

```python
# Slice the dataset in a balanced fashion such that the treatment column has both 0 and 1 values
# Necessary for uplift modeling

treated = df_large[df_large['treatment'] == 1].sample(50_000, random_state = 42)
control = df_large[df_large['treatment'] == 0].sample(50_000, random_state = 42)

df_uplift = pd.concat([treated,control]).sample(frac = 1, random_state = 42)
df_uplift.head()
```

```python
df_uplift.info()
```

```python
df_uplift['treatment'].value_counts()
```

```python
df_uplift.groupby('treatment')['visit'].mean()
```

```python
# Save the new balanced file

df_uplift.to_csv('/Users/rohityadav/Desktop/Git Projects/ml-uplift-modeling-criteo/data/criteo-uplift-v2.1-100K-balanced.csv', index=False)
```
